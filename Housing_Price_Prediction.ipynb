Preprocessing data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.size
```

```python
le = LabelEncoder()
label1 = le.fit_transform(df['mainroad'])
label2 = le.fit_transform(df['guestroom'])
label3 = le.fit_transform(df['basement'])
label4 = le.fit_transform(df['hotwaterheating'])
label5 = le.fit_transform(df['airconditioning'])
label6 = le.fit_transform(df['prefarea'])
label7 = le.fit_transform(df['furnishingstatus'])
```

```python
label1
```

```python
df.drop("mainroad", axis=1, inplace=True)
df.drop("guestroom", axis=1, inplace=True)
df.drop("basement", axis=1, inplace=True)
df.drop("hotwaterheating", axis=1, inplace=True)
df.drop("airconditioning", axis=1, inplace=True)
df.drop("prefarea", axis=1, inplace=True)
df.drop("furnishingstatus", axis=1, inplace=True)

df["Pmainroad"] = label1
df["guestroom"] = label2
df["basement"] = label3
df["hotwaterheating"] = label4
df["airconditioning"] = label5
df["prefarea"] = label6
df["furnishingstatus"] = label7
```

Getting data ready
x = independent variable
y = dependent variable

```python
X = df[['area', 'bedrooms', 'bathrooms', 'stories', 'parking',
       'Pmainroad', 'guestroom', 'basement', 'hotwaterheating',
       'airconditioning', 'prefarea', 'furnishingstatus']]
Y = df['price']
```

```python
X
```

Linear Regression

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()

lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions),bins=50);
```
